# Week 1 — Student Grades Analysis

## Session 4 — pandas DataFrames

```python
import pandas as pd

student_data = {
    "Name": ["Amina", "Omar", "Layla", "Sara"],
    "Grade": [90, 72, 45, 96],
    "Program": [
        "Artificial Intelligence",
        "Software Engineering",
        "Artificial Intelligence",
        "Data Science"
    ]
}

students_df = pd.DataFrame(student_data)
```

```python
print(students_df)
```

```python
print("Shape:", students_df.shape)
print("Columns:", students_df.columns)
print("Index:", students_df.index)
print("Data types:")
print(students_df.dtypes)
```

```python
print(students_df["Name"])
```

```python
print(students_df[["Name", "Grade"]])
```

```python
print(students_df.loc[0])
```

```python
students_df.iloc[2]
```

```python
passing_students = students_df[students_df["Grade"] >= 50]

print(passing_students)
```

```python
ai_students = students_df[
    students_df["Program"] == "Artificial Intelligence"
]

print(ai_students)
```

```python
strong_ai_students = students_df[
    (students_df["Program"] == "Artificial Intelligence")
    & (students_df["Grade"] >= 70)
]

print(strong_ai_students)
```

```python
students_df["Status"] = students_df["Grade"].apply(
    lambda grade: "Pass" if grade >= 50 else "Fail"
)

print(students_df)
```

```python
# use describe() for a quick summary
print(students_df["Grade"].describe())


print("Average grade:", students_df["Grade"].mean())
print("Highest grade:", students_df["Grade"].max())
print("Lowest grade:", students_df["Grade"].min())
```

```python
# count values in a categorical column
print(students_df["Program"].value_counts())

print(students_df["Status"].value_counts())
```

```python
# group by program
program_average = students_df.groupby("Program")["Grade"].mean()

print(program_average)
```

```python
# sort the DataFrame by grade
sorted_students = students_df.sort_values(
    by="Grade",
    ascending=False
)

print(sorted_students)
```

```python
# find the highest and lowest student using pandas

highest_index = students_df["Grade"].idxmax()
lowest_index = students_df["Grade"].idxmin()

highest_student = students_df.loc[highest_index]
lowest_student = students_df.loc[lowest_index]

print(
    f"Highest student: {highest_student['Name']} "
    f"with {highest_student['Grade']}"
)

print(
    f"Lowest student: {lowest_student['Name']} "
    f"with {lowest_student['Grade']}"
)
```

```python
# reset the sorted index
sorted_students = students_df.sort_values(
    by="Grade",
    ascending=False
).reset_index(drop=True)

print(sorted_students)
```

```python
# add a new student row
new_student = {
    "Name": "Adam",
    "Grade": 58,
    "Program": "Artificial Intelligence",
    "Status": "Pass"
}

students_df.loc[len(students_df)] = new_student

print(students_df)
```

```python
# update an existing value
students_df.loc[
    students_df["Name"] == "Adam",
    "Grade"
] = 65

print(students_df)
```

```python
# delete a row
students_df = students_df[
    students_df["Name"] != "Layla"
].reset_index(drop=True)

print(students_df)
```

```python
# check for missing values
print(students_df.isna())

print("Missing values by column:")
print(students_df.isna().sum())
```

```python
# create a missing value
students_df.loc[len(students_df)] = {
    "Name": "Noah",
    "Grade": None,
    "Program": "Data Science",
    "Status": None
}

print(students_df)
```

```python
print(students_df.isna().sum())
```

```python
# fill the missing values
average_grade = students_df["Grade"].mean()

students_df["Grade"] = students_df["Grade"].fillna(average_grade)

students_df["Status"] = students_df["Grade"].apply(
    lambda grade: "Pass" if grade >= 50 else "Fail"
)

print(students_df)
print(students_df.isna().sum())
```

```python
# verify Noha's filled grade
print("Average used:", average_grade)

noah_row = students_df[
    students_df["Name"] == "Noah"
]

print(noah_row)
```

```python
import pandas as pd

project_data = {
    "Name": ["Ali", "Maya", "Noah", "Lina", "Adam"],
    "Grade": [81, 67, None, 93, 58],
    "Program": ["AI", "Software Engineering", "AI", "Data Science", "AI"]
}

project_df = pd.DataFrame(project_data)

# 1. Shape
print(project_df.shape)

# 2. Missing values
print(project_df.isna().sum())

# 3. Fill missing grade with average
avg_grade = project_df["Grade"].mean()
project_df["Grade"] = project_df["Grade"].fillna(avg_grade)

# 4. Add Status column
project_df["Status"] = project_df["Grade"].apply(lambda g: "Pass" if g >= 60 else "Fail")

# 5. Average grade
print(project_df["Grade"].mean())

# 6. Pass/fail counts
print(project_df["Status"].value_counts())

# 7. Highest and lowest student
print(project_df.loc[project_df["Grade"].idxmax()])
print(project_df.loc[project_df["Grade"].idxmin()])

# 8. Group by Program
print(project_df.groupby("Program")["Grade"].mean())

# 9. Sort by grade
print(project_df.sort_values(by="Grade", ascending=False))
```
