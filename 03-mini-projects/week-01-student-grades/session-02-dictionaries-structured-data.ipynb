## Session 2 — Dictionaries and Structured Student Data

```python
student = {
    "name": "Abdul",
    "grade": 85,
    "program": "Artificial Inteligence"
}

print(student)
```

```python
# Access dictionary values
print("Student name: ", student["name"])
print("Student grade: ", student["grade"])
print("Student program: ", student["program"])
```

```python
# Change a dictionary value
print("Original grade: ", student["grade"])

student["grade"] = 80

print("Updated grade: ", student["grade"] )
```

```python
# Add a new key-value pair
student["status"] = "Pass"

print(student)
print("Student status: ", student["status"])
```

```python
# Check whether a key exists
if "status" in student:
    print("The status key exists.")

if "email" not in student:
    print("The email key does not exist.")
```

```python
# Loop through a dictionary
for key in student:
    print(key)

for key, value in student.items():
    print(f"{key}: {value}")
```

```python
# Store multiple students

students = [
    {
        "name": "Abdul",
        "grade": 90,
        "program": "Artificial Intelligence"
    },
    {
        "name": "Omar",
        "grade": 72,
        "program": "Software Engineering"
    },
    {
        "name": "Layla",
        "grade": 45,
        "program": "Artificial Intelligence"
    }
]

print(students)
```

```python
# Print the second student's name:
print("Second student: ", students[1]["name"])
```

```python
# Loop through all students 
for student in students:
    print(f"{student['name']} received a grade of {student['grade']}")
```

```python
# Assign pass or fail dynamically 

for student in students:
    if student["grade"] >= 50:
        status = "Pass"
    else:
        status = "Fail"

    print(f"{student['name']}: {status}")
```

```python
# Add the status into each dictionary 
for student in students:
    if student["grade"] >= 50:
        student["status"] = "Pass"
    else:
        student["status"] = "Fail"

print(students)
```

```python
# Calculate the class average from the dictionaries

total_grade = 0

for student in students:
    total_grade += student["grade"]

class_average = total_grade / len(students)

print(f"Calss average: {class_average:2f}")
```

```python
# Count passing and failing students 

pass_count = 0
fail_count = 0

for student in students:
    if student["grade"] >= 50:
        pass_count += 1
    else:
        fail_count += 1

print("Passing students:", pass_count)
print("Failing students:", fail_count)
```

```python
# Create a reusable summary function
def summarize_students(student_list):
    total_grade = 0
    pass_count = 0
    fail_count = 0

    for student in student_list:
        total_grade += student["grade"]

        if student["grade"] >= 50:
            pass_count += 1
        else:
            fail_count += 1

    class_average = total_grade / len(student_list)

    print(f"Class average: {class_average:.2f}")
    print("Passing students:", pass_count)
    print("Failing students:", fail_count)
```

```python
summarize_students(students)
```

```python
# Protect against an empty list

def summarize_students(student_list):
    if len(student_list) == 0:
        print("No student data available.")
        return

    total_grade = 0
    pass_count = 0
    fail_count = 0

    for student in student_list:
        total_grade += student["grade"]

        if student["grade"] >= 50:
            pass_count += 1
        else:
            fail_count += 1

    class_average = total_grade / len(student_list)

    print(f"Class average: {class_average:.2f}")
    print("Passing students:", pass_count)
    print("Failing students:", fail_count)
```

```python
summarize_students([])
```

```python
# Highest performing student
highest_student = students[0]

for student in students:
    if student["grade"] > highest_student["grade"]:
        highest_student = student

print(
    f"Highest-performing student: "
    f"{highest_student['name']} with {highest_student['grade']}"
)
```

```python
# Find the lowest-performing student
lowest_student = students[0]

for student in students:
    if student["grade"] < lowest_student["grade"]:
        lowest_student = student

print(
    f"Lowest-performing student: "
    f"{lowest_student['name']} with {lowest_student['grade']}"
)
```

```python
# Create reusable functions

def find_highest_student(student_list):
    if len(student_list) == 0:
        return None

    highest_student = student_list[0]

    for student in student_list:
        if student["grade"] > highest_student["grade"]:
            highest_student = student

    return highest_student


def find_lowest_student(student_list):
    if len(student_list) == 0:
        return None

    lowest_student = student_list[0]

    for student in student_list:
        if student["grade"] < lowest_student["grade"]:
            lowest_student = student

    return lowest_student


highest = find_highest_student(students)
lowest = find_lowest_student(students)

print(f"Highest: {highest['name']} with {highest['grade']}")
print(f"Lowest: {lowest['name']} with {lowest['grade']}")
```
