```python
import pandas as pd
import numpy as np

student_data = {
    "Name": ["Ali", "Maya", "Noah", "Lina", "Adam", "Sara"],
    "Python": [82, 67, 49, 93, 58, 88],
    "NumPy": [79, 72, 45, 95, 61, 84],
    "pandas": [85, 69, 52, 91, 56, 90],
    "Program": [
        "AI",
        "Software Engineering",
        "AI",
        "Data Science",
        "AI",
        "Data Science"
    ]
}

students_df = pd.DataFrame(student_data)

print(students_df)
```

```python
students_df["Average"] = students_df[
    ["Python", "NumPy", "pandas"]
].mean(axis=1)

print(students_df)
```

```python
# round the averages
students_df["Average"] = students_df["Average"].round(2)

print(students_df)
```

```python
# add a Status column
students_df["Status"] = students_df["Average"].apply(
    lambda average: "Pass" if average >= 50 else "Fail"
)

print(students_df)
print(students_df["Status"].value_counts())
```

```python
# find the highest and lowest student
highest_index = students_df["Average"].idxmax()
lowest_index = students_df["Average"].idxmin()

highest_student = students_df.loc[highest_index]
lowest_student = students_df.loc[lowest_index]

print(
    f"Highest student: {highest_student['Name']} "
    f"with {highest_student['Average']:.2f}"
)

print(
    f"Lowest student: {lowest_student['Name']} "
    f"with {lowest_student['Average']:.2f}"
)
```

```python
# calculate the average for each subject
subject_averages = students_df[
    ["Python", "NumPy", "pandas"]
].mean()

print(subject_averages.round(2))
```

```python
# group by program
program_averages = students_df.groupby("Program")["Average"].mean()

print(program_averages.round(2))
```

```python
# sort students from highest to lowest average
sorted_students = students_df.sort_values(
    by="Average",
    ascending=False
).reset_index(drop=True)

print(sorted_students)
```

```python
# create a performance category
students_df["Performance"] = students_df["Average"].apply(
    lambda average: (
        "Excellent" if average >= 85
        else "Good" if average >= 70
        else "Needs Improvement"
    )
)

print(students_df[
    ["Name", "Average", "Performance"]
])

print(students_df["Performance"].value_counts())
```

```python
# Final report
final_report = students_df[
    ["Name", "Program", "Average", "Status", "Performance"]
].sort_values(
    by="Average",
    ascending=False
).reset_index(drop=True)

print(final_report)
```

```python
print("Total students:", len(students_df))
print("Class average:", round(students_df["Average"].mean(), 2))
print("Passing students:", (students_df["Status"] == "Pass").sum())
print("Failing students:", (students_df["Status"] == "Fail").sum())

highest_student = students_df.loc[students_df["Average"].idxmax()]
lowest_student = students_df.loc[students_df["Average"].idxmin()]

print(
    f"Highest student: {highest_student['Name']} "
    f"with {highest_student['Average']:.2f}"
)

print(
    f"Lowest student: {lowest_student['Name']} "
    f"with {lowest_student['Average']:.2f}"
)
```
