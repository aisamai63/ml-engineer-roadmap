# Week 1 — Student Grades Analysis

## Session 1 — Python Lists, Loops, Conditions, and Functions

```python
grades = [78, 85,62, 91, 45, 88, 73, 69, 95, 81]
print(grades)
```

```python
print("First grade:", grades[0])
print("Fourth grade:", grades[3])
print("Last grade:", grades[-1])
```

```python
print("Number of grades:", len(grades))
print("Highest grade:", max(grades))
print("Lowest grade:", min(grades))
print("Total of all grades:", sum(grades))
```

```python
total = sum(grades)
number_of_grades = len(grades)

average = total / number_of_grades

print("Average grade:", average)
print(f"Average grade: {average:2f}")
```

```python
for grade in grades:
    print(f"Student grade: {grade}")
```

```python
for grade in grades:
    if grade >= 50:
        print(f"{grade}: Pass")
    else:
        print(f"{grade}: Fail")
```

```python
pass_count = 0

for grade in grades:
    if grade >= 50:
        pass_count += 1
print("Number of passing students:", pass_count)
```

```python
fail_count = 0

for grade in grades:
    if grade < 50:
        fail_count += 1

print("Number of failing students:", fail_count)
```

```python
def calculate_average(values):
    if len(values) == 0:
        return None

    total = sum(values)
    count = len(values)
    average = total / count

    return average
```

```python
class_average = calculate_average(grades)

print(f"Class average: {class_average:f}")
```

```python
# Protect the function from an empty list
calculate_average([])
```

```python
# test the function

empty_result = calculate_average([])
print(empty_result)
```

```python
# Create a passing-count function

def count_passing_grades(values):
    pass_count = 0

    for grade in values:
        if grade >= 50:
            pass_count +=1

    return pass_count
```

```python
number_passing = count_passing_grades(grades)

print("Passing student:", number_passing)
```

```python
def count_failing_grades(values):
    fail_count = 0

    for grade in values:
        if grade < 50:
            fail_count +=1 

    return fail_count
```

```python
number_failing = count_failing_grades(grades)

print("Failing students:", number_failing)
```
