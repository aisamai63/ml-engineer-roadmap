# Week 2 — Personal Expenses Analysis

## Session 2 — Clean Missing and Incorrect Values

```python
import pandas as pd
import numpy as np

expense_data = {
    "Date": [
        "2026-07-01",
        "2026-07-02",
        "2026-07-03",
        "2026-07-04",
        "2026-07-05",
        "2026-07-06",
        "2026-07-06",
        "2026-07-08",
        "2026-07-09"
    ],
    "Category": [
        "Food",
        "transportation",
        "Entertainment",
        "Food ",
        "Utilities",
        "Transportation",
        "Transportation",
        None,
        "Shopping"
    ],
    "Description": [
        "Groceries",
        "Bus pass",
        "Movie ticket",
        "Restaurant",
        "Phone bill",
        "Taxi",
        "Taxi",
        "Coffee",
        "Clothes"
    ],
    "Amount": [
        82.50,
        45.00,
        np.nan,
        37.25,
        65.00,
        22.50,
        22.50,
        -6.75,
        120.00
    ],
    "Payment_Method": [
        "Debit",
        "Credit",
        "credit",
        "Debit ",
        "Debit",
        "Cash",
        "Cash",
        None,
        "Credit"
    ]
}

expenses_df = pd.DataFrame(expense_data)

expenses_df
```

## 1. Inspect the Messy Dataset

```python
print("Dataset shape", expenses_df.shape)
print("\nMissing values:")
print(expenses_df.isna().sum())
print("\nDuplicate rows:")
print(expenses_df.duplicated().sum())
print("\nData types:")
print(expenses_df.dtypes)
```

## 2. Remove Duplicate Rows

```python
expenses_df = expenses_df.drop_duplicates().copy()

print("New shape:", expenses_df.shape)
print("Duplicate rows remaining:", expenses_df.duplicated().sum())
```

## 3. Clean Text Values

```python
expenses_df["Category"] = expenses_df["Category"].str.strip().str.title()
expenses_df["Payment_Method"] = expenses_df["Payment_Method"].str.strip().str.title()

print("Categories:")
print(expenses_df["Category"].unique())

print("\nPayment methods:")
print(expenses_df["Payment_Method"].unique())
```

## 4. Fill Missing Text Values

```python
expenses_df["Category"] = expenses_df["Category"].fillna("Unknown")
expenses_df["Payment_Method"] = expenses_df["Payment_Method"].fillna("Unknown")

print(expenses_df.isna().sum())
```

## 5. Fill the Missing Amount

```python
average_amount = expenses_df["Amount"].mean()

expenses_df["Amount"] = expenses_df["Amount"].fillna(average_amount)

print("Average used: ", round(average_amount, 2))
print("Missing values:")
print(expenses_df.isna().sum())
```

## 6. Find Invalid Amounts

```python
invalid_amounts = expenses_df[expenses_df["Amount"] <= 0]

print(invalid_amounts)
```

## 7. Correct the Invalid Amount

```python
expenses_df.loc[expenses_df["Amount"] < 0, "Amount"] = (
    expenses_df.loc[expenses_df["Amount"] < 0, "Amount"].abs()
)

expenses_df
```

```python
print("Invalid amounts remaining:")
print((expenses_df["Amount"] <= 0).sum())
```

## 8. Convert the Date Column

```python
expenses_df["Date"] = pd.to_datetime(expenses_df["Date"])

print(expenses_df.dtypes)
```

## 9. Reset the Index

```python
expenses_df = expenses_df.reset_index(drop=True)

expenses_df
```

## 10. Final Data Quality Check

```python
print("Dataset shape:", expenses_df.shape)
print("Duplicate rows:", expenses_df.duplicated().sum())
print("Missing values:", expenses_df.isna().sum().sum())
print("Invalid amounts:", (expenses_df["Amount"] <= 0).sum())

print("\nData types:")
print(expenses_df.dtypes)
```
