## 1. Total Spending by Category

```python
category_totals = (
    expenses_df.groupby("Category")["Amount"].sum().sort_values(ascending=False)
)

category_totals
```

## 2. Average Expense by Category

```python
category_averages = (
    expenses_df
    .groupby("Category")["Amount"]
    .mean()
    .round(2)
    .sort_values(ascending=False)
)

category_averages
```

## 3. Count Transactions by Category

```python
category_counts = (
    expenses_df
    .groupby("Category")["Amount"]
    .count()
    .sort_values(ascending=False)
)

category_counts
```

## 4. Spending Summary by Category

```python
category_summary = (
    expenses_df
    .groupby("Category")["Amount"]
    .agg(["count", "sum", "mean", "min", "max"])
    .round(2)
    .sort_values("sum", ascending=False)
)

category_summary
```

## 5. Spending by Payment Method

```python
payment_summary = (
    expenses_df
    .groupby("Payment_Method")["Amount"]
    .agg(["count", "sum", "mean"])
    .round(2)
    .sort_values("sum", ascending=False)
)

payment_summary
```

## 6. Create a Month Column

```python
expenses_df["Month"] = expenses_df["Date"].dt.month_name()

expenses_df[["Date", "Month"]]
```

## 7. Create a Needs or Wants Column

```python
needs_categories = ["Food", "Transportation", "Utilities"]

expenses_df["Needs_or_Wants"] = expenses_df["Category"].apply(
    lambda category: "Need" if category in needs_categories else "Want"
)

expenses_df[["Category", "Needs_or_Wants"]]
```

## 8. Create a Large Expense Column

```python
expenses_df["Large_Expense"] = expenses_df["Amount"].apply(
    lambda amount: "Yes" if amount >= 75 else "No"
)

expenses_df[["Description", "Amount", "Large_Expense"]]
```
