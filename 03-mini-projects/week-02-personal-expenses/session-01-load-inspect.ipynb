# Week 2 — Personal Expenses Analysis

## Session 1 — Load and Inspect Expense Data

```python
import pandas as pd

expense_data = {
    "Date": [
        "2026-07-01",
        "2026-07-02",
        "2026-07-03",
        "2026-07-04",
        "2026-07-05",
        "2026-07-06",
        "2026-07-07",
        "2026-07-08"
    ],
    "Category": [
        "Food",
        "Transportation",
        "Entertainment",
        "Food",
        "Utilities",
        "Transportation",
        "Shopping",
        "Food"
    ],
    "Description": [
        "Groceries",
        "Bus pass",
        "Movie ticket",
        "Restaurant",
        "Phone bill",
        "Taxi",
        "Clothes",
        "Coffee"
    ],
    "Amount": [
        82.50,
        45.00,
        18.00,
        37.25,
        65.00,
        22.50,
        120.00,
        6.75
    ],
    "Payment_Method": [
        "Debit",
        "Credit",
        "Credit",
        "Debit",
        "Debit",
        "Cash",
        "Credit",
        "Cash"
    ]
}

expenses_df = pd.DataFrame(expense_data)

expenses_df
```

## 1. Inspect the Dataset

```python
print("Number of rows:", expenses_df.shape[0])
print("Number of columns:", expenses_df.shape[1])
print("\nColumns names:", expenses_df.columns.tolist())
print(expenses_df.dtypes)
```

## 2. Convert the Date Column

```python
expenses_df["Date"] = pd.to_datetime(expenses_df["Date"])

print(expenses_df.dtypes)
```

## 3. Dataset Information and Summary

```python
expenses_df.info()
```

```python
expenses_df.describe()
```

## 4. Check for Missing Values

```python
expenses_df.isna().sum()
```

## 5. View the First and Last Records

```python
display(expenses_df.head(3))
display(expenses_df.tail(3))
```

## 6. Inspect Unique Values

```python
print("Categories:")
print(expenses_df["Category"].unique())

print("\nPayment methods:")
print(expenses_df["Payment_Method"].unique())
```

## 7. Count Values in Categorical Columns

```python
print("Category counts:")
print(expenses_df["Category"].value_counts())

print("\nPayment method counts:")
print(expenses_df["Payment_Method"].value_counts())
```

## 8. Basic Amount Analysis

```python
print("Total spending:", expenses_df["Amount"].sum())
print("Average expense:", expenses_df["Amount"].mean())
print("Highest expense:", expenses_df["Amount"].max())
print("Lowest expense:", expenses_df["Amount"].min())
```

## 9. Find the Highest Expense Record

```python
highest_expensese_index = expenses_df["Amount"].idxmax()
highest_expense = expenses_df.loc[highest_expensese_index]

highest_expense
```

## 10. Session 1 Summary

```python
print("Dataset shape:", expenses_df.shape)
print("Missing values:", expenses_df.isna().sum().sum())
print("Total spending:", expenses_df["Amount"].sum())
print("Average expense:", round(expenses_df["Amount"].mean(), 2))
print("Most common category:", expenses_df["Category"].mode()[0])
print("Highest expense:", expenses_df["Amount"].max())
```
